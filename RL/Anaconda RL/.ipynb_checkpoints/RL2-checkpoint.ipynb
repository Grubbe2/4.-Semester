```python
class NeuralNet(torch.nn.Module):
    def __init__(self, input_shape, output_shape):
        """
        En neuralt netværksmodel, der består af tre lineære lag med ReLU-aktivationsfunktion.

        Arguments:
            input_shape (int): Størrelsen af inputfunktionen.
            output_shape (int): Størrelsen af outputfunktionen.
        """
        super(NeuralNet, self).__init__()
        # Definerer de lineære lag
        self.linear1 = torch.nn.Linear(input_shape, 150)
        self.linear2 = torch.nn.Linear(150, 100)
        self.linear3 = torch.nn.Linear(100, output_shape)
        # Definerer ReLU-aktivationsfunktionen
        self.activation = torch.nn.ReLU()
    def forward(self, x):
        """
        Fordeler input gennem de forskellige lag i netværket.

        Arguments:
            x (torch.Tensor): Inputtensor til netværket.

        Returns:
            torch.Tensor: Output fra netværket.
        """
         # Fordeler input gennem de lineære lag og aktivationsfunktionen
        x = self.linear1(x)
        x = self.activation(x)
        x = self.linear2(x)
        x = self.activation(x)
        x = self.linear3(x)
        return x
```

```python
def preprocess_state(state, input_shape, add_noice=False):
    """
    Forbehandler en tilstand ved at ændre dens form og muligvis tilføje støj.

    Arguments:
        state (np.ndarray): Tilstanden, der skal forbehandles.
        input_shape (tuple): Formen af det ønskede input til det neurale netværk.
        add_noise (bool, valgfri): Angiver om støj skal tilføjes til tilstanden. Standard er False.

    Returns:
        np.ndarray: Forbehandlet tilstand.
    """
    if add_noice == True: # Hvis add_noise er sand, tilføjes støj til tilstanden
        state = state.reshape(*input_shape) + np.random.rand(*input_shape)/10.0
    else: # Ellers ændres blot formen på tilstanden
        state = state.reshape(*input_shape)
    return state
```

```python
def get_action(Q_values, num_actions, epsilon):
    """
    Vælger en handling baseret på Q-værdier med epsilon-greedy politik.

    Arguments:
        Q_values (np.ndarray): Array af Q-værdier for hver handling.
        num_actions (int): Antal mulige handlinger.
        epsilon (float): Epsilon-værdi, der styrer graden af tilfældighed i valget af handling.

    Returns:
        int: Valgt handling.
    """
    if(random.random() < epsilon): # Hvis en tilfældig værdi er mindre end epsilon, vælges en tilfældig handling
        action = np.random.randint(0,4)
    else: # Ellers vælges handlingen med den højeste Q-værdi
        action = np.argmax(Q_values, 1)[0]
    return action
    
```

```python
class target_maker:
    def __init__(self, terminal_rewards):
        """
        En klasse til at generere mål-værdier for træning af Q-learning.

        Args:
            terminal_rewards (list): En liste over belønninger, der angiver terminaltilstande.
        """
        self.terminal_rewards = terminal_rewards

    def __call__ (self, current_reward, gamma, maxQ):
        """
        Genererer mål-værdier baseret på den aktuelle belønning, rabatfaktor og maksimale Q-værdi.

        Args:
            current_reward (float): Den aktuelle belønning.
            gamma (float): Rabatfaktor, der styrer vægtningen af fremtidige belønninger.
            maxQ (float): Den maksimale forventede fremtidige belønning.

        Returns:
            torch.Tensor: Mål-værdien for Q-learning.
        """
        if current_reward in self.terminal_rewards:
            # Hvis den aktuelle belønning er en terminalbelønning, beregnes målværdien med rabatfaktoren og maksimale Q-værdi
            Y = current_reward + (gamma * maxQ)
        else:
            # Ellers er målværdien blot den aktuelle belønning
            Y = current_reward
        # Konverterer målværdien til en tensor og fjerner den fra grafen for at undgå at påvirke baglæns
        Y = torch.Tensor([Y]).detach()
        return Y
```

```python
input_shape = Grid_size**3
output_shape = len(actions)
# Opretter et neuralt netværk for agenter med inputformen Grid_size**3 og outputformen baseret på antallet af handlinger
Agent = NeuralNet(input_shape, output_shape)
```

```python
# Initialiser variabler og objekter
episodes = 3000  # Antal trænings episoder
epsilon = 1.  # Initial værdi for epsilon i epsilon-greedy politik
gamma = 0.9  # Rabatfaktor for fremtidige belønninger
terminal_rewards = [-10, +10]  # Terminalbelønninger, der angiver afslutningen af et spil
create_target = target_maker(terminal_rewards)  # Opretter en målfunktion baseret på terminalbelønninger
loss_fn = torch.nn.MSELoss()  # Definerer tabsevnen til at være Mean Squared Error
learning_rate = 0.001  # Læringshastigheden for optimizeren
optimizer = torch.optim.Adam(Agent.parameters(), lr=learning_rate)  # Definerer Adam optimizeren med det neurale netværks parametre og læringshastigheden

# Start træningens hovedløkke
for i in range(episodes):
    env = Gridworld(Grid_size=Grid_size, mode='static')  # Opretter et nyt gridverdenmiljø
    print(i)  # Udskriver det nuværende trænings-episode nummer

    # Initialiser starttilstanden for den aktuelle episode
    state1 = env.state
    state1 = preprocess_state(state1, (1, input_shape), add_noice=True)  # Forbehandler starttilstanden
    state1 = torch.from_numpy(state1).float()  # Konverterer starttilstanden til en tensor

    # Start episode-loop
    gate = 1
    while (gate == 1):
        Q_val = Agent(state1)  # Beregner Q-værdier for de tilgængelige handlinger i den aktuelle tilstand
        Q_val_numpy = Q_val.detach().numpy()  # Konverterer Q-værdierne til en numpy-array
        action = get_action(Q_val_numpy, len(actions), epsilon)  # Vælger en handling baseret på epsilon-greedy politik

        # Udfør handlingen og observer de resulterende tilstande og belønninger
        state2, reward, done, _ = env.step(action)
        state2 = preprocess_state(state2, (1, input_shape), add_noice=True)  # Forbehandler de nye tilstande
        state2 = torch.from_numpy(state2).float()  # Konverterer de nye tilstande til en tensor
        env.render()  # Vis miljøet

        # Opdater Q-værdierne baseret på de nye observationer
        with torch.no_grad():
            newQ = Agent(state2)
        maxQ = torch.max(newQ)
        Y = create_target(reward, gamma, maxQ)  # Genererer målværdien for Q-learning
        X = Q_val.squeeze()[action]
        loss = loss_fn(X, Y)  # Beregner tabet
        optimizer.zero_grad()  # Nulstiller gradienterne
        loss.backward()  # Beregner gradienterne
        optimizer.step()  # Opdaterer vægtene

        # Opdaterer tilstanden til den nye tilstand
        state1 = state2

        # Check om episode er afsluttet
        if done:
            gate = 0

    # Opdater epsilon for epsilon-greedy politik for at mindske graden af tilfældighed over tid
    if epsilon > 0.1:
        epsilon -= (1 / episodes)

    # Luk gridverdenmiljøet
    env.close()
```

# Train On Batches

```python
def get_batch(replay, batch_size, device = 'cpu'):
    minibatch = random.sample(replay, batch_size)
    state1_batch = torch.cat([s1 for (s1,a,r,s2,d) in minibatch]).to(device)
    action_batch = torch.Tensor([a for (s1,a,r,s2,d) in minibatch]).long().to(device)
    reward_batch = torch.Tensor([r for (s1,a,r,s2,d) in minibatch]).to(device)
    state2_batch = torch.cat([s2 for (s1,a,r,s2,d) in minibatch]).to(device)
    done_batch = torch.Tensor([d for (s1,a,r,s2,d) in minibatch]).to(device)
    return state1_batch, action_batch, reward_batch, state2_batch, done_batch
    
```

```python
def batch_target(reward_batch, gamma, maxQ, done_batch):
    Y = reward_batch + (gamma * maxQ * (1 - done_batch))
    return Y
```

```python
def update(loss, optimizer):
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
```

```python
def reward_metric(all_rewards_list, last_n_reward = 20):
    if len(all_rewards_list) >= last_n_reward:
        mean_reward = np.mean(all_rewards_list[-last_n_reward])
    else: 
        mean_reward = 0
    return mean_reward
```

```python
from collections import deque
memory_size = 5000# Størrelsen af hukommelsesbufferen
batch_size = 200
replay = deque(maxlen=memory_size)
device = 'cuda'
# Sætter træningen til at foregå på GPU'en for hurtigere læring

Agent = NeuralNet(input_shape, output_shape).to(device)
# Opretter en instans af neuralt netværksmodellen og flytter den til den angivne enhed (GPU)

episodes = 3000
epsilon = 1.
gamma = 0.9
losses = [] 
terminal_rewards = [-10, +10]
create_target = target_maker(terminal_rewards)
# Opsætning af hyperparametre og nødvendige datastrukturer til træning

loss_fn = torch.nn.MSELoss()
learning_rate = 0.001
optimizer = torch.optim.Adam(Agent.parameters(), lr=learning_rate)
# Definerer tabfunktionen (Mean Squared Error) og optimizeren (Adam) til træning af det neurale netværk

all_rewards_list = []
mean_rewards = []
# Lister til at gemme belønninger og gennemsnitlige belønninger over tid til monitorering af fremskridt

for i in range(episodes):
    env = Gridworld(Grid_size=Grid_size, mode='random', frame_size=(800,800,3))
    env.reset()
    state1 = env.state
    state1 = preprocess_state(state1,(1, input_shape), add_noice=True)
    state1 = torch.from_numpy(state1).float().to(device)
    # Initialiserer miljøet og forbehandler den indledende tilstand

    gate = 1
    while (gate == 1):
        # Løkke til interaktion med miljøet indtil afslutning
        Q_val = Agent(state1)
        Q_val_numpy = Q_val.detach().cpu().numpy()
        action = get_action(Q_val_numpy, len(actions), epsilon)
        # Får handling fra agentens politik (epsilon-greedy strategi)

        state2, reward, done, _ = env.step(action)
        state2 = preprocess_state(state2,(1, input_shape), add_noice=True)
        state2 = torch.from_numpy(state2).float().to(device)
        # Udfører handling i miljøet og forbehandler den resulterende tilstand

        exp = (state1, action, reward, state2, done)
        replay.append(exp)
        # Gemmer erfaringstuple i genafspilningshukommelsen

        if len(replay) > batch_size:
            # Hvis genafspilningshukommelsen er stor nok, begynd at træne det neurale netværk
            state1_batch, action_batch, reward_batch, state2_batch, done_batch = get_batch(replay, batch_size, device)
            Q1 = Agent(state1_batch)

            with torch.no_grad():
                Q2 = Agent(state2_batch)
            maxQ = torch.max(Q2, axis=1)[0]

            Y = batch_target(reward_batch, gamma, maxQ, done_batch)
            # Beregner mål-Q-værdier til træning

            X = Q1.gather(dim=1, index=action_batch.unsqueeze(dim=1)).squeeze()
            loss = loss_fn(X, Y)
            losses.append(loss.item())
            # Beregner tab og gemmer det

            update(loss, optimizer)
            # Opdaterer det neurale netværksparametre

            all_rewards_list.append(reward)
            mean_reward = reward_metric(all_rewards_list, 100)
            mean_rewards.append(mean_reward)
            # Sporer belønninger og gennemsnitlige belønninger over tid

        state1 = state2
        if done == True:
            gate = 0
        # Opdaterer nuværende tilstand og kontrollerer afslutning

    if epsilon > 0.1:
        epsilon -= (1 / episodes)
    # Aftagende epsilon for epsilon-greedy politik

    env.close()
    # Lukker miljøet
```

```python
from collections import deque
import copy

# Størrelsen af hukommelsesbufferen
memory_size = 5000

# Batch-størrelse til træning
batch_size = 200

# Opretter en deque til at fungere som en genafspilningshukommelse
replay = deque(maxlen=memory_size)

# Sætter træningen til at foregå på GPU'en for hurtigere læring
device = 'cuda'

# Opsætning af hyperparametre til træning
episodes = 3000
epsilon = 1.
gamma = 0.9

# Definerer belønninger for terminaltilstande
terminal_rewards = [-10, +10]

# Opretter en funktion til at generere målværdier
create_target = target_maker(terminal_rewards)

# Definerer tabfunktionen (Mean Squared Error) til træning af det neurale netværk
loss_fn = torch.nn.MSELoss()

Agent = NeuralNet(input_shape, output_shape).to(device)
target_NN = copy.deepcopy(Agent)
target_NN.load_state_dict(Agent.state_dict())

sync_freq = 500
k = 0

# Definerer læringshastigheden for Adam-optimizeren
learning_rate = 0.001

# Opretter en Adam-optimizer og forbinder den med agentens parametre
optimizer = torch.optim.Adam(Agent.parameters(), lr=learning_rate)

# Lister til at gemme belønninger og gennemsnitlige belønninger over tid til monitorering af fremskridt
all_rewards_list = []
mean_rewards = []

# Loop over episoderne
for i in range(episodes):
    # Initialisering af miljøet
    env = Gridworld(Grid_size=Grid_size, mode='random', frame_size=(800,800,3))
    env.reset()
    state1 = env.state
    
    # Forbehandling af den indledende tilstand
    state1 = preprocess_state(state1, (1, input_shape), add_noice=True)
    state1 = torch.from_numpy(state1).float().to(device)

    # Flag til at styre loopet
    gate = 1
    
    # Loop over handlinger indtil afslutning
    while (gate == 1):
        k = k + 1
        # Beregning af Q-værdier for den aktuelle tilstand
        Q_val = Agent(state1)
        Q_val_numpy = Q_val.detach().cpu().numpy()
        
        # Valg af handling baseret på epsilon-greedy strategi
        action = get_action(Q_val_numpy, len(actions), epsilon)

        # Udførelse af handling i miljøet
        state2, reward, done, _ = env.step(action)
        
        # Forbehandling af den resulterende tilstand
        state2 = preprocess_state(state2, (1, input_shape), add_noice=True)
        state2 = torch.from_numpy(state2).float().to(device)

        # Gemmer erfaringstuple i genafspilningshukommelsen
        exp = (state1, action, reward, state2, done)
        replay.append(exp)

        # Træning af det neurale netværk hvis genafspilningshukommelsen er stor nok
        if len(replay) > batch_size:
            state1_batch, action_batch, reward_batch, state2_batch, done_batch = get_batch(replay, batch_size, device)
            Q1 = Agent(state1_batch)

            with torch.no_grad():
                Q2 = target_NN(state2_batch)
            
            maxQ = torch.max(Q2, axis=1)[0]
            Y = batch_target(reward_batch, gamma, maxQ, done_batch)

            X = Q1.gather(dim=1, index=action_batch.unsqueeze(dim=1)).squeeze()
            loss = loss_fn(X, Y)
            losses.append(loss.item())

            # Opdatering af det neurale netværk
            update(loss, optimizer)

            # Sporing af belønninger og gennemsnitlige belønninger over tid
            all_rewards_list.append(reward)
            mean_reward = reward_metric(all_rewards_list, 100)
            mean_rewards.append(mean_reward)
        
        # Opdatering af nuværende tilstand og kontrol af afslutning
        state1 = state2

        if k % sync_freq == 0:
            target_NN.load_state_dict(Agent.state_dict())
        
        if done == True:
            gate = 0
    
    # Opdatering af epsilon for epsilon-greedy politik
    if epsilon > 0.1:
        epsilon -= (1 / episodes)
    
    # Lukning af miljøet
    env.close()
```

```python
def resize_and_bgr2gray(image):
    image = image[0:288, 0:404]
    image_data = cv2.cvtColor(cv2.resize(image, (84,84)), cv2.COLOR_BGR2GRAY)
    image_data[image_data > 0] = 255
    image_data = np.reshape(image_data, (84,84,1))
    return image_data
```

```python
class Conv_Q_Net(torch.nn.Module):
    def __init__(self, input_depth_length, output_shape):
        super(Conv_Q_Net, self).__init__()
        self.conv1 = torch.nn.Conv2d(input_depth_length, 32, 8, stride=4)
        self.conv2 = torch.nn.Conv2d(32, 64, 4, stride=2)
        self.conv3 = torch.nn.Conv2d(64, 64, 3, stride=1)
        self.linear1 = torch.nn.Linear(3136, 512)
        self.linear2 = torch.nn.Linear(512, output_shape)
        self.flatten = torch.nn.Flatten()
        self.activation = torch.nn.ReLU()

    def forward(self, x): 
        x = self.conv1(x)
        x = self.activation(x)
        x = self.conv2(x)
        x = self.activation(x)
        x = self.conv3(x)
        x = self.activation(x)
        x = self.flatten(x)
        x = self.linear1(x)
        x = self.activation(x)
        x = self.linear2(x)
        return x
        
```
